```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("Salary_dataset.csv")
data.head()
```

```python
# View summary statistics
data.describe().T
```

```python
# Visualize the data
plt.figure(figsize=(5, 5))
sns.scatterplot(x='YearsExperience', y='Salary', data=data)
```

# Recherche manuelle de w et b

```python
# Trying to fit a line manually
w = 10000
b = 21000
y = w * data.YearsExperience + b
plt.figure(figsize=(5, 5))
plt.scatter(data.YearsExperience, data.Salary, label='Data')
plt.plot(data.YearsExperience, y, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Calculate Mean Squared Error
mse = np.mean((data.Salary - y) ** 2)/2
print(f'Mean Squared Error: {mse}')
```

```python
# Predicting salary for a new experience value
def predict(X):
    return w * X + b
```

# Implémenter la descente de gradient pour la recherche automatique

```python
def gradient_descent(X, Y, w, b, learning_rate):
    N = len(X)
    y_pred = w * X + b
    dw = (1/N) * np.sum(X * (y_pred - Y))
    db = (1/N) * np.sum(y_pred - Y)
    w -= learning_rate * dw
    b -= learning_rate * db
    return w, b
```

```python
# Gradient Descent Loop
while True:
    w, b = gradient_descent(data.YearsExperience, data.Salary, w, b, learning_rate=0.01)
    y = w * data.YearsExperience + b
    mse_temp = mse
    mse = np.mean((data.Salary - y) ** 2)/2
    print(f'Mean Squared Error: {mse}')
    if abs(mse_temp - mse) == 0: # Convergence atteinte
        print("\nConvergence atteinte.\n")
        print(f'Poids final (w): {w}')
        print(f'Biais final (b): {b}')
        break
    if mse < 1000000: # Seuil d'arrêt pour éviter les itérations infinies
        break
```
